**Linear Regression**

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/drive/MyDrive/intership/spark/data/task1.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
x=data.iloc[:,:-1]
x.head()
```

```python
y=data.iloc[:,1:]
y.head()
```

```python
# Plotting the distribution of scores
data.plot(x='Hours', y='Scores', style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y,
                            test_size=0.2, random_state=0)
```

**Model Creation**

```python
#model creation
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
plt.scatter(X_train,y_train,color="Blue")
plt.plot(np.array(X_train),reg.predict(X_train),color="red")
```

 **Making Predictions**
****

```python
print(X_test)
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

**Evaluating the model**

```python
from sklearn import metrics
print('Mean Absolute Error:',
      metrics.mean_absolute_error(y_test, y_pred))
```

```python
#Predicted score for student who studies for 9.25 hrs/day
hours = np.array(9.25)
own_pred = reg.predict(hours.reshape(1, -1))
```

```python
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0][0]))
```
